**UNIVARIATE ANALYSIS**

```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# 2. Load the original dataset
df = catalog.load("scoring_data_raw")

# 3. Set style
sns.set(style="whitegrid")

# 4. Select numeric and categorical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# 5. Create a Univariate Summary Table for Numeric Features
summary_data = []
for col in numeric_cols:
    data = df[col].dropna()
    summary_data.append({
        'Feature': col,
        'Mean': round(data.mean(), 2),
        'Std Dev': round(data.std(), 2),
        'Min': round(data.min(), 2),
        'Max': round(data.max(), 2),
        'Missing (%)': round(df[col].isnull().mean() * 100, 2),
        'Skewness': round(skew(data), 2)
    })

univariate_summary = pd.DataFrame(summary_data)

# 6. Display the summary table
from IPython.display import display
display(univariate_summary)

# 7. Plot Histograms with KDE for Numeric Features
for col in numeric_cols:
    plt.figure(figsize=(10, 4))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# 8. Plot Countplots for Categorical Features
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Frequency of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

BIVARIATE ANALYSIS

```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu
from IPython.display import display

# 2. Load Dataset
df = catalog.load("scoring_data_raw")

# 3. Define Target and Columns
target = 'BAD'
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.drop(target, errors='ignore')
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# 4. Pearson Correlation Matrix (Numeric vs Numeric)
plt.figure(figsize=(12, 8))
corr = df[numeric_cols.tolist() + [target]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix (Numeric Variables)")
plt.tight_layout()
plt.show()

# 5. Boxplots (Numeric vs Target)
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=target, y=col, data=df)
    plt.title(f'{col} by {target}')
    plt.tight_layout()
    plt.show()

# 6. Stacked Bar Charts (Categorical vs Target)
for col in categorical_cols:
    if df[col].nunique() <= 10:  # Skip if too many categories
        crosstab = pd.crosstab(df[col], df[target], normalize='index') * 100
        crosstab.plot(kind='bar', stacked=True, figsize=(8, 4), colormap='tab20c')
        plt.title(f'Stacked Bar Plot: {col} vs {target}')
        plt.ylabel('Percentage')
        plt.legend(title='BAD')
        plt.tight_layout()
        plt.show()

# 7. Mann-Whitney U Test (Numeric vs Target)
mann_whitney_results = []
for col in numeric_cols:
    group0 = df[df[target] == 0][col].dropna()
    group1 = df[df[target] == 1][col].dropna()
    if len(group0) > 0 and len(group1) > 0:
        stat, p = mannwhitneyu(group0, group1, alternative='two-sided')
        mann_whitney_results.append({'Feature': col, 'U-statistic': round(stat, 2), 'p-value': round(p, 5)})

mann_whitney_df = pd.DataFrame(mann_whitney_results).sort_values(by='p-value')
display(mann_whitney_df)
```
