### PCA

```python
df = final_data.copy()
```

```python
# Identify objective variable (last column assumed)
objective_var = 'BAD'

# Separate features and target
X = df.drop(columns=[objective_var])
y = df[objective_var]
```

```python
# Select only numeric features
X_numeric = X.select_dtypes(include=[np.number])

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)
```

```python
# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Explained variance
explained_var = pca.explained_variance_ratio_
print("Explained variance per component:", explained_var)
```

```python
# Scree plot
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(explained_var), marker='o')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Variance Growth:**

As expected, the cumulative explained variance increases as the number of PCA components increases. This indicates that each additional component captures more variability in the data.

**Diminishing Returns:**

After approximately 15 components, the rate of increase in explained variance slows down. This suggests diminishing returns in the contribution of additional components to explaining the variability.

**Importance:**

The first eight components explain around 60% of the variability in the data. Notably, the first three components account for approximately 30% of the explained variance in the dataset.

### Correlation With Objective Variable

```python
# Calculate correlations between PCs and the objective variable
correlations = []
for i in range(X_pca.shape[1]):
    corr = np.corrcoef(X_pca[:, i], y)[0, 1]
    correlations.append(corr)
```

```python
# Define PC indices
pc_indices = [f"PC{i+1}" for i in range(len(correlations))]

# Sort correlations and their corresponding PCs in descending order
sorted_indices = np.argsort(correlations)[::-1]
sorted_correlations = [correlations[i] for i in sorted_indices]
sorted_pc_indices = [pc_indices[i] for i in sorted_indices]

# Plot the sorted correlations
plt.figure(figsize=(12, 7))
plt.bar(sorted_pc_indices, sorted_correlations, color='skyblue', edgecolor='black')
plt.axhline(0, color='gray', linewidth=0.7, linestyle='--')

# Adding labels and title
plt.title('Correlation of Principal Components with Objective Variable (BAD)', fontsize=14)
plt.xlabel('Principal Components', fontsize=12)
plt.ylabel('Correlation Coefficient', fontsize=12)
plt.grid(axis='y', linestyle='--', color='lightgray', alpha=0.7)

plt.show()
```

```python
# Correlation with objective variable
obj_corr = corr_matrix[objective_var].drop(objective_var).sort_values(ascending=False)
print("\nCorrelation of each variable with objective variable:")
print(obj_corr)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=obj_corr.values, y=obj_corr.index, color='skyblue')
plt.title(f'Correlation of Each Variable with {objective_var}')
plt.xlabel('Correlation')
plt.ylabel('Variables')
plt.tight_layout()
plt.show()
```

### Others + Circle

```python
predictors_scaled = X_scaled
target = y
explained_variance = pca.explained_variance_ratio_
print("Explained Variance by Principal Components:", explained_variance)
    
# Calculate correlation between PCs and target variable
pca_scores = pd.DataFrame(pca.transform(predictors_scaled))
pca_scores["UnitSales"] = target.values
correlations = pca_scores.corr()["UnitSales"][:-1]
correlation_rank = correlations.abs().sort_values(ascending=False)
print("Correlation with Target Variable (Ranked):")
print(correlation_rank)
```

```python
predictors = X
# Identify contribution of each variable to PCs
loadings = pd.DataFrame(pca.components_.T, columns=[f"PC{i+1}" for i in range(pca.n_components_)], index=predictors.columns)
    
# Calculate overall contributions
weighted_contributions = (loadings.iloc[:, 0].abs() * correlation_rank.iloc[0]).sort_values(ascending=False)
contributions = pd.DataFrame({
        "Variable": weighted_contributions.index,
        "Weighted_Contribution": weighted_contributions.values
    }).sort_values(by="Weighted_Contribution", ascending=False)
print("Variable Contributions to Target Variability:")
print(contributions)
```

```python
# Plot contributions
plt.figure(figsize=(10, 6))
sns.barplot(data=contributions, x="Weighted_Contribution", y="Variable", palette="mako")
plt.title("Variable Contributions to Target Variability")
plt.xlabel("Weighted Contribution")
plt.ylabel("Variables")
plt.show()
```

**Top Contributing Variables:**
MRTDUE, VALUE, and CLNO have the highest contributions, suggesting that these variables explain a significant portion of the variance or are strongly associated with the primary target variable. These likely represent key financial factors in the dataset.

**Mid-Level Contributors:**
Variables such as JOB_Other, JOB_ProfExe, LOAN, and CLAGE show moderate contributions. These are less dominant than the top three but still important in understanding variability.

**Lower-Level Contributors:**
Variables like VALUE_missing, DELINQ, and YOJ have minimal contributions, suggesting limited influence on explaining variability. They might be less relevant in the model or contribute noise.

**Categorical Variables:**
Categories like JOB_Other, REASON_DebtCon, and REASON_HomeImp provide insight into their importance relative to financial or demographic data, implying distinct patterns in job type or purpose of financial decisions.

**Actionable Insights:**
Variables with high contributions, like MRTDUE and VALUE, should be prioritized for interpretation and further analysis. Lower-contributing variables might be candidates for exclusion if reducing dimensionality is desired.

**Target Focus:**
The dominance of financial indicators (e.g., MRTDUE, VALUE) suggests a model heavily driven by economic factors.

```python
# PCA - Variable Correlation Circle
plt.figure(figsize=(8, 8))
for i, (x, y) in enumerate(zip(loadings.iloc[:, 0], loadings.iloc[:, 1])):
        plt.arrow(0, 0, x, y, head_width=0.02, head_length=0.02, color='lightblue')
        plt.text(x*1.1, y*1.1, loadings.index[i], color='navy', size = 'xx-small')
plt.title("PCA - Variable Correlation Circle")
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid()
plt.show()
```

**PC1 (Horizontal Axis):**
Variables like VALUE, MRTDUE, and CLNO have high positive loadings, indicating strong relationships with PC1.
REASON_HomeImp shows a high negative loading, suggesting it contrasts with the variables on the right of PC1.

**PC2 (Vertical Axis):**
Variables like REASON_DebtCon and DELINQ_missing have strong positive loadings on PC2, while JOB_Other has a strong negative loading.

**Clusters of Variables:**
JOB_ProfExe, CLAGE, and VALUE form a cluster, showing they might be associated with similar trends in the dataset.

Missing values such as DELINQ_missing and DEROG_missing are clustered together, possibly highlighting missing data patterns.

**Variables Near the Origin:**
Variables like YOJ, DEROG, and JOB_Self contribute less to the variation explained by PC1 and PC2.

**Interpretation of Principal Components:**
PC1 could represent a contrast between financial stability (high mortgage due, high value, and more loans) versus lower-value or improvement-related financial activities (e.g., REASON_HomeImp).
PC2 might capture missing information patterns (DELINQ_missing, DEROG_missing) or distinct job categories (JOB_Other vs. JOB_ProfExe).
